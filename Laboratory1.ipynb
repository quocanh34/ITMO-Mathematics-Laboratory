# Laboratory 1: The Riemann Integral 

The given function: $y = 3^x$ with the interval [1,2]

Task 1: Make integral sum for the given function.

Task 2: Calculate the integral through the limit of integral sums. Check it using Newton-Leibniz formula.

Task 3: Write a program to visualize the Riemann sums.

Task 4: Calculate the estimation error, compare it with theoretical value.

For more information: [laboratory link](https://drive.google.com/file/d/1chy5HUKgyP8UhWa9kQD7wlZbVGkpAYs9/view?usp=sharing)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
a = 1; b = 2; N = 25
n = 10

gap = b-a
theoretical_value = 6 / (np.log(3))

# Given function
f = lambda x : 3 ** x

# Function to calculate the error
def estimation_error(n, category):
  riemann_sum = 0 
  if category == "left":
    for i in range(0,n):
      x_current = a + gap*i/N
      riemann_sum += (gap/N) * f(x_current)
    return abs(round(riemann_sum - theoretical_value, 5))

  if category == "middle":
    for i in range(0,n):
      x_current = a + gap*i/N
      riemann_sum += (gap/N) * f((2*x_current + gap/n)/2)
    return abs(round(riemann_sum - theoretical_value, 5))

  if category == "right":
    for i in range(0,n):
      x_current = a + gap*i/N
      riemann_sum += (gap/N) * f(x_current + gap/n)
    return abs(round(riemann_sum - theoretical_value, 5))


x = np.linspace(a,b,N+1)
y = f(x)

# Use n*N+1 points to plot the function smoothly
X = np.linspace(a,b,n*N+1) 
Y = f(X)

plt.figure(figsize=(20,8))

# Left Riemann Sum
plt.subplot(1,3,1)
plt.plot(X,Y,'r')
x_left = x[:-1] 
y_left = y[:-1]
plt.plot(x_left,y_left,'b.',markersize=5)
plt.bar(x_left,y_left,width=gap/N,alpha=0.2,align='edge',edgecolor='b')
plt.title('Left Riemann Sum, N = {}'.format(N))
plt.text(1.0, 8, r'Estimation error: ' + str(estimation_error(len(x)-1, "left")))


# Midpoint Riemann Sum
plt.subplot(1,3,2)
plt.plot(X,Y,'r')
x_mid = (x[:-1] + x[1:])/2
y_mid = f(x_mid)
plt.plot(x_mid,y_mid,'b.',markersize=5)
plt.bar(x_mid,y_mid,width=gap/N,alpha=0.2,edgecolor='b')
plt.title('Midpoint Riemann Sum, N = {}'.format(N))
plt.text(1.0, 8, r'Estimation error: ' + str(estimation_error(len(x)-1, "middle")))

# Right Riemann Sum
plt.subplot(1,3,3)
plt.plot(X,Y,'r')
x_right = x[1:] 
y_right = y[1:]
plt.plot(x_right,y_right,'b.',markersize=5)
plt.bar(x_right,y_right,width=-gap/N,alpha=0.2,align='edge',edgecolor='b')
plt.title('Right Riemann Sum, N = {}'.format(N))
plt.text(1.0, 8, r'Estimation error: ' + str(estimation_error(len(x), "right")))

plt.show()
```

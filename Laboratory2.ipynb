```python
import matplotlib.pyplot as plt
from numpy import *
import numpy as np
```

```python
def get_an(n):
  if n == 0:
    return (1 + 5 / 2 * np.pi) / (3 * np.pi)
  return (-6 * n + 9 * np.sin(np.pi * n / 3)) / (np.pi * n * (4 * n ** 2 - 9))

def get_bn(n):
  return (-4 * n ** 2 + 9 - 9 * np.cos(np.pi * n / 3)) / (np.pi * n * (4 * n ** 2 - 9))

def get_sum(n, x):
  s = get_an(0)
  if n == 1:
    return np.zeros_like(x) + s
  for i in range(1, n):
    s += get_an(i) * np.cos(i * 2 * x / 3) + get_bn(i) * np.sin(i * 2 * x / 3)
  return s

plt.figure(figsize=(20,8))

#sum=10
plt.subplot(1,3,1)
plt.plot(np.linspace(0, np.pi / 2), np.sin(np.linspace(0, np.pi / 2)), c="red", label=r"$f(x)$")
plt.plot(np.linspace(np.pi / 2, 3 * np.pi), [1] * 50, c="red")
num1 = 10
num_str = str(10)
plt.plot(np.linspace(0, 3 * np.pi, 500), get_sum(num1, np.linspace(0, 3 * np.pi, 500)), c="green", label=fr"$N=10$")
plt.grid(True)
plt.legend()

#sum=50
plt.subplot(1,3,2)
plt.plot(np.linspace(0, np.pi / 2), np.sin(np.linspace(0, np.pi / 2)), c="red", label=r"$f(x)$")
plt.plot(np.linspace(np.pi / 2, 3 * np.pi), [1] * 50, c="red")
num2 = 50
num_str = str(50)
plt.plot(np.linspace(0, 3 * np.pi, 500), get_sum(num2, np.linspace(0, 3 * np.pi, 500)), c="green", label=fr"$N=50$")
plt.grid(True)
plt.legend()

#sum=100
plt.subplot(1,3,3)
plt.plot(np.linspace(0, np.pi / 2), np.sin(np.linspace(0, np.pi / 2)), c="red", label=r"$f(x)$")
plt.plot(np.linspace(np.pi / 2, 3 * np.pi), [1] * 50, c="red")
num3 = 100
num_str = str(100)
plt.plot(np.linspace(0, 3 * np.pi, 500), get_sum(num3, np.linspace(0, 3 * np.pi, 500)), c="green", label=fr"$N=100$")
plt.grid(True)
plt.legend()

```
